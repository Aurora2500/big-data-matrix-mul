```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
python = pd.read_csv('results/python.csv')
java = pd.read_csv('results/java.csv')
c_cached = pd.read_csv('results/cached_c.csv')
c_naive = pd.read_csv('results/naive_c.csv')
c_threaded = pd.read_csv('results/threaded_c.csv')
rust = pd.read_csv('results/rust_wgpu.csv')
```

```python
xlab = python['Size']

plt.plot(python['Time'], label='python')
plt.plot(java['Time'], label='java')
plt.plot(c_naive['Time'], label='c naive')
plt.plot(c_cached['Time'], label='c cached')
plt.plot(c_threaded['Time'], label='c threaded')
plt.plot(rust['Time'], label='rust')
plt.xticks(range(12), labels=xlab)
plt.xlabel("Matrix size")
plt.ylabel("Time (s)")
plt.legend()
plt.show()
```

```python
xlab = python['Size']

def rescale(x):
	return np.log10(x+1e-6)

yticks = [
	f"$10^{{{s}}}$"
	for s in range(-6, 5, 2)
]

plt.plot(rescale(python['Time']), label='python')
plt.plot(rescale(java['Time']), label='java')
plt.plot(rescale(c_naive['Time']), label='c naive')
plt.plot(rescale(c_cached['Time']), label='c cached')
plt.plot(rescale(c_threaded['Time']), label='c threaded')
plt.plot(rescale(rust['Time']), label='rust')
plt.xticks(range(12), labels=xlab)
plt.yticks(range(-6, 5, 2), labels=yticks)
plt.ylim(-6.5, 4.5)
plt.xlabel("Matrix size")
plt.ylabel("Time (s)")
plt.hlines(range(-6, 5, 2), 0, 11, linestyle="dashed", colors='black', linewidth=0.4)
plt.legend()
plt.show()
```
